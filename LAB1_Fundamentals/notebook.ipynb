### 1.2

```python
peppers = io.imread('peppers.png')
print(peppers.ndim)
print(peppers.shape)

# extract the red component
red_peppers = peppers[:,:,0]
print(red_peppers.ndim)
print(red_peppers.shape)
plt.imshow(red_peppers, cmap='gray')
plt.show()

# G and B to 0
green_blue_zero = peppers.copy()
green_blue_zero[:,:,1] = 0
green_blue_zero[:,:,2] = 0
plt.imshow(green_blue_zero)
plt.show()
```

## Exercise 2

### 2.1

```python
peppers_hsv = skimage.color.rgb2hsv(peppers)
plt.imshow(peppers_hsv)
plt.show()

# extract the hue channel
hue = peppers_hsv[:,:,0]
plt.imshow(hue, cmap='gray')
plt.show()

# extract the saturation channel
saturation = peppers_hsv[:,:,1]
plt.imshow(saturation, cmap='gray')
plt.show()

# extract the value channel
value = peppers_hsv[:,:,2]
plt.imshow(value, cmap='gray')
plt.show()
```

### 2.1


```python
# mosaic 192 x 192 x 3
mosaic = np.zeros((192, 192, 3), dtype=np.uint8)

mosaic[0:64, 0:64, :] = 0 # black
mosaic[0:64, 64:128, 0] = 255 # red
mosaic[0:64, 128:192, 2] = 255 # blue
mosaic[64:128, 0:64, 1] = 255 # green
mosaic[64:128, 64:128, :2] = 255 # yellow
mosaic[64:128, 128:192, ::2] = 255 # magenta
mosaic[128:192, 0:64, 1:] = 255 # cyan
mosaic[128:192, 64:128, :] = 255 # white
mosaic[0:64, 0:64, :] = 0 # black
plt.imshow(mosaic)
plt.show()

```

```python
plt.bar(bins, h)
```

```python
# Re-bin into 32 groups
h_grouped = h.reshape(32, 8).sum(axis=1)

# Make new bin edges (0255 split into 32 parts)
bin_edges = np.linspace(0, 256, 32+1)[:-1]  # start of each bin

plt.bar(bin_edges, h_grouped)
```

```python
# Read the image
img = io.imread("187_0070.jpg")

# Extract channels
R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]

# --- Red channel ---
plt.subplot(2, 3, 1)
plt.imshow(R, cmap="Reds")
plt.title("Red Channel")
plt.axis("off")

plt.subplot(2, 3, 4)
plt.hist(R.ravel(), bins=256, color="red")
plt.title("Red Histogram")

# --- Green channel ---
plt.subplot(2, 3, 2)
plt.imshow(G, cmap="Greens")
plt.title("Green Channel")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.hist(G.ravel(), bins=256, color="green")
plt.title("Green Histogram")

# --- Blue channel ---
plt.subplot(2, 3, 3)
plt.imshow(B, cmap="Blues")
plt.title("Blue Channel")
plt.axis("off")

plt.subplot(2, 3, 6)
plt.hist(B.ravel(), bins=256, color="blue")
plt.title("Blue Histogram")

plt.tight_layout()
plt.show()
```
